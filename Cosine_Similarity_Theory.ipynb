**Cosine Similarity :**
  - It is a measurement that quantifies the similarity between two or more vectors. i.e. It is the cosine angle between vectors, which are typically non-zero and within an inner product space.
  - It is described mathamatically as the division between the dot product of vectors and the product of the euclidean norms/magnitude of each vector.
